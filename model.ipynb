```python
import pandas as pd
```

```python
df=pd.read_csv("netflix_titles.csv")
```

```python
df=df[["type", "title", "director", "cast", "listed_in", "description"]]
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df["director"].value_counts()
```

```python
def Latter(l):
    new_word = ""
    
    for i in l:          
        if i != " ":     
            new_word += i
    
    return [new_word]
```

```python
df["title"]=df["title"].apply(Latter)
df["director"]=df["director"].apply(Latter)
df["cast"]=df["cast"].apply(Latter)
df["listed_in"]=df["listed_in"].apply(Latter)
df["description"]=df["description"].apply(Latter)
df["type"]=df["type"].apply(Latter)
```

```python
df.head()
```

```python
df["tags"]=df["director"]+df["cast"]+df["listed_in"]+df["description"]
```

```python
df.head()
```

```python
new=df.drop(columns=["director", "cast","listed_in", "description" ])
```

```python
new.head()
```

```python
new["tags"]=new["tags"].apply (lambda x:" ".join(x))
```

```python
new["title"]=new["title"].apply (lambda x:" ".join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000, stop_words="english" )
```

```python
vector=cv.fit_transform(new["tags"]).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim=cosine_similarity(vector)
```
